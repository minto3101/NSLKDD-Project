```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df = pd.read_csv("../input/nslkdd-with-labels/KDDTrain-_20Percent.csv")
df
```

```python
testdf = pd.read_csv("../input/nslkdd-with-labels/KDDTest(With Labels).csv")
testdf
```

```python
df = df.drop(columns= ["Protocol Type","Service","Flag"])
testdf = testdf.drop(columns= ["Protocol Type","Service","Flag"])
```

```python
df.loc[df["Class"] == "normal" , "Class"] = 0
df.loc[df["Class"] != 0 , "Class"] = 1
testdf.loc[testdf["Class"] == "normal" , "Class"] = 0
testdf.loc[testdf["Class"] != 0 , "Class"] = 1
```

```python
column_list = list(df)
column_list[-2], column_list[-1] = column_list[-1], column_list[-2]
column_list
```

```python
df = df[column_list]
df["Class"] = df.Class.astype(int)
testdf = testdf[column_list]
testdf["Class"] = testdf.Class.astype(int)
```

```python
x_train = df.iloc[:,:-1]
x_test = testdf.iloc[:,:-1]

y_train = df.iloc[:,-1]
y_test = testdf.iloc[:,-1]

x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
