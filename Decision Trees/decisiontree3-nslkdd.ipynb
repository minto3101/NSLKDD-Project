```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
traindf = pd.read_csv("../input/nslkdd-with-labels/Full Train for Classification (with strings).csv")
traindf = traindf.drop(columns=["Unnamed: 0", "Service"])
traindf
```

```python
testdf = pd.read_csv("../input/nslkdd-with-labels/Test Dataset for Classification (with strings).csv")
testdf = testdf.drop(columns="Service")
testdf
```

```python
#Applying Label Encoders
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
```

```python
traindf["Protocol Type"] = le1.fit_transform(traindf["Protocol Type"])
traindf["Flag"] = le2.fit_transform(traindf["Flag"])

testdf["Protocol Type"] = le1.fit_transform(testdf["Protocol Type"])
testdf["Flag"] = le2.fit_transform(testdf["Flag"])
```

```python
testdf
```

```python
df = pd.concat([testdf,traindf])
df
```

```python
x_train = df.iloc[22544:,:-1]
y_train = df.iloc[22544:,-1]

x_test = df.iloc[:22544,:-1]
y_test = df.iloc[:22544,-1]
```

```python
y_train
```

```python
#Model without PCA 
```

```python
from sklearn import tree
model_NoPCA = tree.DecisionTreeClassifier(random_state=16)
model_NoPCA.fit(x_train,y_train)
model_NoPCA.score(x_test,y_test)
```

```python
#Model with PCA 
```

```python
x = df.iloc[:,:-1]
```

```python
from sklearn.decomposition import PCA
maxi=0
for i in range(41,0,-1):
    pca1 = PCA(n_components=i)
    x = pca1.fit_transform(x)
    
    x_train = x[22544:,:]
    x_test = x[:22544,:]
    
    model_PCA = tree.DecisionTreeClassifier(random_state=16)
    model_PCA.fit(x_train,y_train)
    if (model_PCA.score(x_test,y_test)>maxi):
        maxi = model_PCA.score(x_test,y_test)
        best_para = i
    
```

```python
print(maxi)
print(best_para)
```
