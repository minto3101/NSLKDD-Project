```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("../input/nslkdd-with-labels/Full Train for Classification (with strings).csv")
df = df.drop(columns=["Unnamed: 0","Service"])
df
```

```python
testdf = pd.read_csv("../input/nslkdd-with-labels/Test Dataset for Classification (with strings).csv")
testdf = testdf.drop(columns="Service")
testdf
```

```python
subdf = pd.read_csv("../input/nslkdd-with-labels/20Subset Train for Classification (with strings).csv")
subdf = subdf.drop(columns="Service")
subdf
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
```

```python
df["Protocol Type"] = le1.fit_transform(df["Protocol Type"])
df["Flag"] = le2.fit_transform(df["Flag"])
```

```python
testdf["Protocol Type"] = le1.fit_transform(testdf["Protocol Type"])
testdf["Flag"] = le2.fit_transform(testdf["Flag"])
```

```python
subdf["Protocol Type"] = le1.fit_transform(subdf["Protocol Type"])
subdf["Flag"] = le2.fit_transform(subdf["Flag"])
```

```python
x_train = df.iloc[:,:-1]
x_test = testdf.iloc[:,:-1]
xsub_train = subdf.iloc[:,:-1]

y_train = df.iloc[:,-1]
y_test = testdf.iloc[:,-1]
ysub_train = subdf.iloc[:,-1]
```

```python
from sklearn import tree
modelFull = tree.DecisionTreeClassifier(random_state = 1)
modelFull.fit(x_train,y_train)
maxi = modelFull.score(x_test,y_test)
maxi
```

```python
for i in range(100):
    model_i = tree.DecisionTreeClassifier(random_state=i)
    model_i.fit(x_train,y_train)
    if (model_i.score(x_test,y_test)>maxi):
        maxi = model_i.score(x_test,y_test)
        best_para = i
```

```python
print(maxi)
print(best_para)
```

```python
modelSub = tree.DecisionTreeClassifier(random_state = 1)
modelSub.fit(xsub_train,ysub_train)
maxi = modelSub.score(x_test,y_test)
maxi
```

```python
for i in range(100):
    model_i = tree.DecisionTreeClassifier(random_state=i)
    model_i.fit(xsub_train,ysub_train)
    if (model_i.score(x_test,y_test)>maxi):
        maxi = model_i.score(x_test,y_test)
        best_para = i
```

```python
print(maxi)
print(best_para)
```
