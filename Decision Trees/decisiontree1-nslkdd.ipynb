```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("../input/nslkdd-with-labels/Full Train for Classification (without strings).csv")
df = df.drop(columns= "Unnamed: 0")
df
```

```python
testdf = pd.read_csv("../input/nslkdd-with-labels/Test Dataset for Classification (without strings).csv")
testdf
```

```python
x_train = df.iloc[:,:-1]
x_test = testdf.iloc[:,:-1]

y_train = df.iloc[:,-1]
y_test = testdf.iloc[:,-1]
```

```python
x_test
```

```python
from sklearn import tree
model1 = tree.DecisionTreeClassifier(random_state = 1)
```

```python
model1.fit(x_train,y_train)
```

```python
maxi = model1.score(x_test,y_test)
maxi
```

```python
for i in range(100):
    model_i = tree.DecisionTreeClassifier(max_features=39, random_state=i)
    model_i.fit(x_train,y_train)
    if (model_i.score(x_test,y_test)>maxi):
        maxi = model_i.score(x_test,y_test)
        best_para = i
    
```

```python
print(maxi)
print(best_para)
```
