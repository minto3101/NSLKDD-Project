```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
#Performing Label Encoding for string type features and then applying Logistic Regression
```

```python
df = pd.read_csv("../input/nslkdd-with-labels/20Subset Train for Classification (with strings).csv")
testdf = pd.read_csv("../input/nslkdd-with-labels/Test Dataset for Classification (with strings).csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
df["Protocol Type"] = le1.fit_transform(df["Protocol Type"])
df["Service"] = le2.fit_transform(df["Service"])
df["Flag"] = le3.fit_transform(df["Flag"])
```

```python
testdf["Protocol Type"] = le1.fit_transform(testdf["Protocol Type"])
testdf["Service"] = le2.fit_transform(testdf["Service"])
testdf["Flag"] = le3.fit_transform(testdf["Flag"])
```

```python
x_train = df.iloc[:,:-1]
x_test = testdf.iloc[:,:-1]

y_train = df.iloc[:,-1]
y_test = testdf.iloc[:,-1]
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

```python
x_train = x_train.drop(columns="Service")
x_test = x_test.drop(columns="Service")
```

```python
x_test
```

```python
model2 = LogisticRegression()
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```
